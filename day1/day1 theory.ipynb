## Sets

```python
our_work = set(["metacognition", "mindwandering", "perception"])
print our_work
```

```python
our_work[0]
```

```python
our_work.add("consciousness")
print our_work
```

```python
our_work.add("consciousness")
print our_work
our_work.add("consciousness")
print our_work
```

```python
competing_labs_work = set(["motor control", "decision making", "memory", "consciousness"])
what_we_should_focus_on = our_work - competing_labs_work
print what_we_should_focus_on
```

```python
what_we_should_avoid = our_work.intersection(competing_labs_work)
print what_we_should_avoid
```

## Ifs and breaks

```python
a = 2
b = 5
c = a + b 
if c < 6:
    print "yes"
```

```python
if c < 6:
    print "yes"
else:
    print "no"
```

```python
if c < 6:
    print "yes"
elif c > 6:
    print "no"
else:
    print "kind of"
```

```python
if True:
    print "true, true!"
```

```python
if False:
    print "never!"
```

```python
for i in range(10):
    if i == 6:
        break
    print i
```

```python
for i in range(10):
    if i == 6:
        continue
    print i
```
